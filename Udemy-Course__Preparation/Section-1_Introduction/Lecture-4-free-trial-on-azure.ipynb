# Lecture 4. Free trial on Azure



## What to note
  - If you have a cloud account, you can use it to register for a 14-day Databricks full trial.

  - A full trial in your cloud is recommended 
  since it includes production grade functionalities 
  that are not available in the limited community edition.

  - Microsoft offers a free tier for 12 months, 
  allowing you to explore and try out Azure services 
  free of charge and up to specified limits for each service.



## Steps
  - You can create a Databricks workspace directly through Azure Portal.

    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure portal.jpg" style="width:960px" >
    </div>

  - In the search bar, search for "Databricks" and from Services, select Azure Databricks.

    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure portal search Databricks.jpg" style="width:960px" >
    </div>

  - Now click create.
  
    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure Databricks create.jpg" style="width:960px" >
    </div>

  - Create an Azure Databricks workspace

    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Create an Azure Databricks workspace.jpg" style="width:960px" >
    </div>

    * Select your subscription
    * Create a resource group: `DatabricksDemoRG`
    * Add a friendly name: `DemoWorkspace`
    * Select a location: `West Europe`
    * Select pricing tier
      There are two main options: Standard and the Premium. 
      Premium tier includes all features of standard tier as well as Role-based access control. 
      For the free trial, choose the **14-day free premium option**.
    * Now click "Review and Create".

      <div style="text-align: center;">
        <img src="../../assets/images/Screen-Captures/Create an Azure Databricks workspace - review.jpg" style="width:960px" >
      </div>

    * Click "Create". The deployment takes some minutes.

      <div style="text-align: center;">
        <img src="../../assets/images/Screen-Captures/Create an Azure Databricks workspace - Deployment in progress.jpg" style="width:960px" >
      </div>

    * After deployment is completed, click "Go to resource".

      <div style="text-align: center;">
        <img src="../../assets/images/Screen-Captures/Create an Azure Databricks workspace - Deployment complete.jpg" style="width:960px" >
      </div>

  - Launch the Workspace
  
    * Before launching the workspace, let us take a look on the managed resource group.

      <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure Databricks workspace overview - focus Managed Resource Group.jpg" style="width:960px" >
      </div>
    
      If you navigate to this resource group, you will see some resources here. They are the data plan resources.

      <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure Databricks workspace - Managed Resource Group.jpg" style="width:960px" >
      </div>

    It contains a virtual network, a network security group for managing the inbound and outbound traffic, and a storage account, which is the underlying storage for DBFS.
  
  * Let us come back to our resource and the click "Launch Workspace". 

    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure Databricks workspace - Launch Workspace.jpg" style="width:960px" >
    </div>

    Notice here that it is using as your Active Directory single sign on to login to Databricks platform.
    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Active Directory single sign on to login to Databricks platform.jpg" style="width:960px" >
    </div>

    Here is our Azure Databricks workspace dashboard.
    <div style="text-align: center;">
      <img src="../../assets/images/Screen-Captures/Azure Databricks platform.jpg" style="width:960px" >
    </div>
